```python
sql = """
SELECT 
    ProductID,CampaignID,
    AVG(Likes) AS AvgLikes,
    AVG(Views) AS AvgViews,
    AVG(Clicks) AS AvgClicks,
    ROUND(SUM(Likes) * 100.0 / (SELECT SUM(Likes) FROM engagement_data), 2) AS Likes_Percentage,
    ROUND(SUM(Views) * 100.0 / (SELECT SUM(Views) FROM engagement_data), 2) AS Views_Percentage,
    ROUND(SUM(Clicks) * 100.0 / (SELECT SUM(Clicks) FROM engagement_data), 2) AS Clicks_Percentage,
    ROUND(
        (SUM(Likes) * 100.0 / (SELECT SUM(Likes) FROM engagement_data)) +
        (SUM(Views) * 100.0 / (SELECT SUM(Views) FROM engagement_data)) +
        (SUM(Clicks) * 100.0 / (SELECT SUM(Clicks) FROM engagement_data))
    , 2) AS Total_Percentage
FROM engagement_data
GROUP BY ProductID, CampaignID
ORDER BY Total_Percentage ASC
LIMIT 5;
"""
```

```python
curr.execute(sql)
```

# How do customer reviews impact the purchasing behaviour?

### Analysis & Proposed Solutions: Impact of Reviews on Customer Purchase Behavior 

### **Solution:**  

1. **Low-Rated Products (Avg Rating ≤ 3.5)** – Product IDs: `7, 4, 12, 9, 16, 10`  
   - **Key Concern:** These products have received the lowest ratings, indicating quality issues, unmet customer expectations, or poor performance.  
   - **Solution:** Businesses should analyze the reviews for these products to identify recurring complaints. Possible actions include improving product quality, adjusting pricing, enhancing features, or offering better customer support. Encouraging satisfied customers to leave positive reviews can also help improve overall ratings.  

2. **Moderate-Rated Products (3.5 - 4 Avg Rating)** – Product IDs: `2, 13, 3, 6, 17`  
   - **Key Concern:** These products have decent ratings but still show room for improvement. Customers may have minor issues affecting their overall satisfaction.  
   - **Solution:** Identifying and addressing specific customer concerns, such as minor defects, packaging, or missing features, can help boost ratings. Promotional campaigns, better after-sales support, and targeted improvements based on feedback can lead to higher customer satisfaction.  

3. **High-Rated Products (Avg Rating > 4)** – Product IDs: `18, 15, 1, 11, 20, 5, 19, 8`  
   - **Key Concern:** These products are well-received with strong customer satisfaction. Maintaining this level of quality and reputation is crucial.  
   - **Solution:** Businesses should continue to promote these products, leveraging their positive reviews to drive more sales. Ensuring consistent quality, providing excellent after-sales support, and using customer testimonials in marketing can help sustain high ratings.  

4. **Customers Who Rated Below 3 (Total: 38 Customers)**  
   - **Issue Identified:** A total of **38 customers** have given a rating below **3**, which suggests dissatisfaction with specific products.  
   - **Solution:** Businesses should analyze these ratings in relation to the affected products. If certain products frequently receive low ratings, they may need immediate improvement or repositioning in the market. Personalized outreach to these customers with discounts or replacements can also help recover brand trust.  

5. **Mismatch Between Review Text and Rating (Total: 11 Cases)**  
   - **Issue Identified:** **11 instances** were found where customers mentioned a **5-star rating** in their review text but provided a lower numerical rating.  
   - **Potential Causes:** Possible reasons include **user confusion**, **technical errors**, or **misalignment between review text and rating selection**.  
   - **Solution:** Improving the review submission UI, adding a confirmation step, and using **sentiment analysis** to detect inconsistencies can help resolve this issue. Businesses can also manually verify and correct such discrepancies where applicable.

```python
sql = """
SELECT 
    CASE 
        WHEN ReviewText LIKE '%bad%' OR ReviewText LIKE '%Average%' OR ReviewText LIKE '%performance%' THEN 'Performance Issue'
        WHEN ReviewText LIKE '%Not worth%' OR ReviewText LIKE '%not%' OR ReviewText LIKE '%stopped%'  THEN 'Quality Issue'
        WHEN ReviewText LIKE '%cheaper%' OR ReviewText LIKE '%problem%' OR ReviewText LIKE '%Product did not meet my expectations.%'THEN 'Average'
        WHEN ReviewText LIKE '%late%' Then 'Delivery Issue'
        WHEN ReviewText LIKE '%unclear%' Then 'Unclear Instruction'
        ELSE 'Other Issues'
    END AS ComplaintCategory,
    count(*) AS ComplaintCount
FROM cust_reviews
WHERE Rating <= 3
GROUP BY ComplaintCategory
ORDER BY ComplaintCount DESC;
"""
```

```python
curr.execute(sql)
```
